```python
df1.rename(columns={
    'PatientId': 'Patient_id',
    'AppointmentID': 'Appointment_id',
    'ScheduledDay': 'Scheduled_day',
    'AppointmentDay': 'Appointment_day',
    'Neighbourhood': 'Neighbourhood',
    'Hipertension': 'Hypertension',
    'Handcap': 'Handicap',
    'SMS_received': 'SMS_received',
    'No-show': 'No_show'
}, inplace=True)
```

```python
df1.info()
```

```python
df1['Patient_id'] = df1['Patient_id'].astype('int64')
df1['Scheduled_day'] = pd.to_datetime(df1['Scheduled_day']).dt.date.astype('datetime64[ns]')
df1['Appointment_day'] = pd.to_datetime(df1['Appointment_day']).dt.date.astype('datetime64[ns]')
df1['Scheduled_day'] = df1['Scheduled_day'].dt.strftime('%d-%m-%Y')
df1['Appointment_day'] = df1['Appointment_day'].dt.strftime('%d-%m-%Y')
```

```python
df1.sample(10)
```

```python
df1.Age.value_counts()
```

```python
print("Before change")
print("Patients with Age less than -1 -> {}".format(df1[df1.Age == -1].shape[0]))
print("Patients with Age equal to 0 -> {}".format(df1[df1.Age == 0].shape[0]))
print("Patients with Age greater than 110 -> {}".format(df1[df1.Age > 110].shape[0]))

# Keep only valid ages
df1 = df1[(df1["Age"] >= 0) & (df1["Age"] <= 110)]

print(df1["Age"].value_counts())

print("After change")
print("Patients with Age less than -1 -> {}".format(df1[df1.Age == -1].shape[0]))
print("Patients with Age equal to 0 -> {}".format(df1[df1.Age == 0].shape[0]))
print("Patients with Age greater than 110 -> {}".format(df1[df1.Age > 110].shape[0]))
```

```python
df1["No_show"] = df1["No_show"].replace({"Yes": 1, "No": 0})
```
